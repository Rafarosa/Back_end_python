## 3\. Strings

Aplique três métodos distintos na *string* abaixo, você pode conferir alguns métodos neste [link](https://www.w3schools.com/python/python_ref_string.asp):

```
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'

###Converter texto para minúsculo 
convert_lower = cancao.lower()
print(convert_lower)

### Quantas vezes aparece a palavra Roda
roda = convert_lower.count('roda')
print(f'A palavra roda aparece {roda} no texto')

### trocando a palavra Roda por Quadrado
roda_por_quadrado = convert_lower.replace('roda', 'Quadrado')
print(roda_por_quadrado)
```

Extraia da string abaixo o valor da taxa **selic** na variável `selic` e o valor do **ano** na variavel `ano`. Imprima os valores na tela.

```
noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'
local_selic = noticia.find('%')
selic = noticia[local_selic-4:local_selic]
print(f'O valor da selic é de {selic}')

local_ano = noticia.find('anos')
ano = noticia[local_ano-2]
print(f'O valor da variavel ano é {ano}')
```
